```python
def get_next_symbol(lines):
    counter = 0
    for l in lines:
        if len(l.rstrip().lstrip()) > 0:
            if l.startswith('>"') or l.startswith('"<'):
                return counter
            else:
                counter += 1
    return counter
        
    
```

```python
def format_text(lines, text, line, symb_count, i, sid):

    if len(lines) >= 1:  
        if len(lines[0].strip()) == 0:
            return format_text(lines[1:], text+'\n', line, symb_count, 1, 1)
        
        symb_count = get_next_symbol(lines)   
        #print(symb_count, ' ' , lines)
        if symb_count == 0:
            Nsymb_count = get_next_symbol(lines[1:])
            #print('Ns' , Nsymb_count, ' ' , lines)
            if Nsymb_count > 1:
                startIdx = sid
                endIdx = sid + Nsymb_count - 1
                line = str(startIdx) + '-' + str(endIdx) + '\t' +lines[0].rstrip().lstrip()[2:-2]+'\n'
                text += line
                return format_text(lines[1:], text, line, 0, 1, sid)
            else:
                #sid += 1
                line = str(sid) + '\t' +lines[0].strip()[2:-2] + '\t'
                text += line
                
                return format_text(lines[1:], text, line, symb_count, 1, sid)
            
        elif symb_count == 1:            
            if lines[0].startswith('#'):
                line = '\n'+lines[0]+'\n' 
                text += line
                return format_text(lines[1:], text, line, symb_count, 1, 1)
            else:
                line = '\t' + lines[0].rstrip().lstrip() + ' \t - \t - \t - \t - \t - \n'
                text += line
                sid += 1
            
            return format_text(lines[1:], text, line, symb_count, 1, sid)
      
        elif symb_count > 1:
            for j in range(0, symb_count):
                if len(lines[j].rstrip().lstrip()) > 0:
                    line = str(sid) + '\t' + lines[j].rstrip().lstrip() + ' \t - \t - \t - \t - \t - \n'
                    text += line
                    sid += 1      
            return format_text(lines[symb_count:], text, line, symb_count, i, sid)
        
        else:
            return format_text(lines[1:], text, line, symb_count, 1, sid)
        
    

    return text, symb_count, i

lines = ["##sfd", '"<the>"', "is", "a", "sentence",  '\n', '"<a>"',"new", "finally",'"<endxxx>"', "end", "\n"]
text, symb_count, i = format_text(lines, text='', line='', symb_count=0, i=0, sid=1)
print('text: \n%s\nsymbol count:%s\ni: %d' %(text, symb_count, i))
```
